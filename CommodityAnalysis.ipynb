```python
plt.figure(figsize=(20,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Lumber"])
plt.xticks(rotation="vertical")
plt.title("Lumber Quart Price from 2005 ")
plt.xlabel("Date")
plt.ylabel("Average Price Quar")



plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.xticks(rotation="vertical")
plt.scatter(comm_hou_df["DATE"], comm_hou_df["Lumber"], marker="o", color="red")
```

```python
plt.figure(figsize=(20,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Steel"])
plt.xticks(rotation="vertical")
plt.title("Steel Quart Price from 2005 ")
plt.xlabel("Date")
plt.ylabel("Average Price Quart")



plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Copper"])
plt.xticks(rotation="vertical")
plt.title("Copper Quart Price from 2005 ")
plt.xlabel("Date")
plt.ylabel("Average Price Quart")



plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["WTI Oil"])
plt.xticks(rotation="vertical")
plt.title("WTI oil Quart Price from 2005 ")
plt.xlabel("Date")
plt.ylabel("Average Price Quart")



plt.show()
```

```python
plt.figure(figsize=(15,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "yellow")
plt.plot(comm_hou_df["DATE"],comm_hou_df["WTI Oil"],label='West Texas Intermediate (WTI), Dollars per Barrel',color= "blue")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Copper"], label="Producer Price Index Copper Wire and Cable",color= "black")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Steel"],label="Producer Price Index Iron and Steel", color= "red")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Lumber"],label= "Producer Price Index Lumber", color= "green")

plt.xticks(rotation="vertical")
plt.title("Housing Index Quart Price from 2005 ")
plt.xlabel("Date")
plt.ylabel("Average Price Quart")


plt.legend(loc="best")
plt.show()
```
