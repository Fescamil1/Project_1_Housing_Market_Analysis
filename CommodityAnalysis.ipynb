```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
# %matplotlib notebook
```

```python
Path_comm_hou="Resources/Commodity_Housing_Quarterly_1995.csv"
comm_hou_df=pd.read_csv(Path_comm_hou)
comm_hou_df.head()
```

```python
# analysis_df=commodity_price_df.merge(oil_price_df, on ="DATE",how="outer")
# analysis_df=analysis_df.merge(Hosuing_Ind_df,on="DATE",how="outer")

# analysis_df.head (10)
# # analysis_df.dtypes
# # analysis_df.DATE[0]
```

```python
plt.figure(figsize=(30,10))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Lumber"],label= "Producer Price Index Lumber", color= "green")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "black")
plt.xticks(rotation="vertical",size=10)
plt.title("Lumber Quart Price from 1995 ",size=20)
plt.xlabel("Date (Quarter)",size=20)
plt.ylabel("Average Price Quarter",size=20)
plt.legend(loc="best",fontsize=20)
plt.savefig("commodity_Lumber.png",transparent=True)
plt.show()
```

```python
plt.figure(figsize=(30,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Steel"],label="Producer Price Index Iron and Steel", color= "red")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "black")
plt.xticks(rotation="vertical",size=10)
plt.title("Steel Quart Price from 1995 ",size=20)
plt.xlabel("Date (Quarter)",size=20)
plt.ylabel("Average Price Quart",size=20)
plt.legend(loc="best",fontsize=20)
plt.savefig("commodity_Steel.png",transparent=True)
plt.show()
```

```python
plt.figure(figsize=(30,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "yellow")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Copper"], label="Producer Price Index Copper Wire and Cable",color= "black")
plt.xticks(rotation="vertical",size=10)
plt.title("Copper Quart Price from 1995 ",size=20)
plt.xlabel("Date (Quarter)",size=20)
plt.ylabel("Average Price Quarterly",size=20)
plt.legend(loc="best",fontsize=20)
plt.savefig("commodity_copper.png",transparent=True)
plt.show()
```

```python
plt.figure(figsize=(30,5))
plt.plot(comm_hou_df["DATE"],comm_hou_df["WTI Oil"],label='West Texas Intermediate (WTI), Dollars per Barrel',color= "blue")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "yellow")
plt.xticks(rotation="vertical",size=10)
plt.title("WTI oil Quart Price from 1995 ",size=20)
plt.xlabel("Date (Quarters)",size=20)
plt.ylabel("Average Price Quarterly",size=20)
plt.legend(loc="best",fontsize=20)
plt.savefig("commodity_oil.png",transparent=True)
plt.show()
```

```python
plt.figure(figsize=(30,10))
plt.plot(comm_hou_df["DATE"],comm_hou_df["Hosing Index"], label="All-Transactions House Price Index for Texas",color= "yellow")
plt.plot(comm_hou_df["DATE"],comm_hou_df["WTI Oil"],label='West Texas Intermediate (WTI), Dollars per Barrel',color= "blue")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Copper"], label="Producer Price Index Copper Wire and Cable",color= "black")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Steel"],label="Producer Price Index Iron and Steel", color= "red")
plt.plot(comm_hou_df["DATE"],comm_hou_df["Lumber"],label= "Producer Price Index Lumber", color= "green")
plt.xticks(rotation="vertical")
plt.title("Housing Index Quart Price from 1995 ",size=15)
plt.xlabel("Date",size=15)
plt.ylabel("Average Price Quart",size=15)
plt.legend(loc="best",fontsize=15)
plt.vlines(comm_hou_df.DATE.loc[comm_hou_df.DATE=="1/1/2008"],0,500)
plt.vlines(comm_hou_df.DATE.loc[comm_hou_df.DATE=="4/1/2009"],0,500)
plt.vlines(comm_hou_df.DATE.loc[comm_hou_df.DATE=="1/1/2020"],0,500)
plt.savefig("commodity_collective.png",transparent=True)
plt.show()
```
